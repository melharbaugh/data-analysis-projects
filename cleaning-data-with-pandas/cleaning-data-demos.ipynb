```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv("student_records_dirty.csv")
```

```python
students.head()
```

```python
students.shape
```

```python
students.sample(5)
```

```python
students['course_grade'].value_counts(dropna=False)
```

```python
students.isna().sum()
```

```python
students['course_grade'] = students['course_grade'].replace('N/A', np.nan)
```

```python
students.duplicated().sum()
#checks for duplicated rows
students = students.drop_duplicates()
#removes duplicated rows
```

```python
students['first_name'].str.strip()
```

```python
students['major'] = students['major'].str.strip().str.capitalize()
students['major'] = students['major'].replace({
'Comp sci' : 'Computer science',
'Cs' : 'Computer science',
'Bio' :  'Biology',
'Psych' : 'Psychology'
})
```

```python
students['major'].value_counts()
```

```python
students['course_grade'] = students['course_grade'].str.strip()
students.loc[students['course_grade'] == ' ninety', 'course_grade'] = 90
```

```python
students['age'].value_counts(dropna=False)
students['age'] = pd.to_numeric(students['age'], errors = 'coerce')
```

```python
students['age'].describe()
```

```python
students.loc[students['age'] < 5, 'age'] = np.nan
students.loc[students['age'] > 120, 'age'] = np.nan
```

```python
#create a column combining first and last names
students['full_name'] = students['first_name'] + ' ' + students['last_name']
students['full_name']
```

```python
students['age_group'] = pd.cut(
    students['age'],
    bins=[0, 25, 40, 65], 
    labels=['Young Adults', 'Working Adults', 'Retired']
    )
```

```python
students[['age','age_group']].value_counts()
```

```python
students['cost_per_semester'] = students['tuition_paid'] / 3
```

```python
students['cost_per_semester']
```

```python
students['enrollment_date'] = pd.to_datetime(students['enrollment_date'])
```

```python
students.to_csv('../student_reports_clean.csv', index=False)
#save clean data as new csv
```
