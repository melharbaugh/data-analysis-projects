```python
import matplotlib.pyplot as plt
```

Bar Chart

```python
plants = ['lemon', 'lime', 'Orange']
fruit_yeild = [89, 43,12]

plt.bar(plants, fruit_yeild)
```

Scatter Plot

```python
butterflies=[12,34,89]
pollinator_participants=[10,56,102]

plt.scatter(pollinator_participants, butterflies)
```

Histogram: 
Need a bin size. Bins are the columns. 

```python
butterflies=[2,6,17,28,59,112,56,24]

#counted number of butterflies for 8 weeks = 8 bins
plt.hist(butterflies, bins=8)
```

Pie Chart

```python
categories = ["annuals", "perennials", "trees", "houseplants"]
sales = [678, 1894, 1234, 768]

plt.pie(sales,labels=categories)
```
