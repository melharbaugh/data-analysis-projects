Below we set up the dataframe with the Goodreads dataset for you. If you run this cell, you may get an error that some lines were skipped. Do not worry about that! For this studio, we want to focus on creating an effective visualization.

```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a 
#better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

With your dataframe set up, time to start creating visualizations!

```python
#text ratings count vs reviews count
plt.figure(figsize=(8,6))
plt.scatter(df['text_reviews_count'], df['ratings_count'], alpha=0.6)
plt.title('Text Reviews vs Ratings')
plt.xlabel('Text Reviews')
plt.ylabel('Ratings')
plt.show()
```

```python
# Books published in different Languages 
language_counts = df['language_code'].value_counts()
```

```python
language_counts.index
```

```python
language_counts.values
```

```python
plt.bar(language_counts.index, language_counts.values, color='purple')
plt.title('Book Languages')
plt.xlabel('Language Codes')
plt.ylabel('Count')
plt.show()
```
